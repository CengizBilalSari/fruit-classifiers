```python
import pandas as pd
import os
# 2. Mount Google Drive
from google.colab import drive
import pandas as pd
import numpy as np
drive.mount('/content/drive')
MY_DRIVE_PATH = "/content/drive/MyDrive/MLProject"
DATA_FOLDER = os.path.join(MY_DRIVE_PATH, 'Data google sheet')
PROCESSED_CSV_FILE = os.path.join(DATA_FOLDER, 'Processed_Fruits_Data.csv')
ONEHOT_CSV_FILE = os.path.join(DATA_FOLDER, 'One_Hot_Processed_Fruits_Data.csv')
```

```python
processed_df = pd.read_csv(PROCESSED_CSV_FILE, sep = ";")
# Converting categorical attributes to one hot
categorical_columns = ['Season', 'Origin', 'Color']
processed_df = pd.get_dummies(processed_df, columns=categorical_columns, dtype='uint8')
processed_df.to_csv(ONEHOT_CSV_FILE, index=False, sep=';', encoding='utf-8-sig')
```
